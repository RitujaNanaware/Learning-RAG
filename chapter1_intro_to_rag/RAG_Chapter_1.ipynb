```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
```

```python
texts = [
    "RAG enhances LLMs with external knowledge.",
    "Embedding models convert text into vectors.",
    "FAISS is used for fast vector search.",
    "LangChain helps in building RAG pipelines.",
    "Large Language Models are powerful but hallucinate."
]

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(texts)

print("Shape of embeddings:", embeddings.shape)
```

```python
# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# Sample query
query = "How does RAG use vector search?"
query_embedding = model.encode([query])

# Search
k = 2
D, I = index.search(query_embedding, k)

print("Top", k, "results:")
for idx in I[0]:
    print(f"- {texts[idx]}")
```

This notebook demonstrates the core idea of RAG: converting documents into embeddings, storing them in a vector store, and retrieving relevant content based on query similarity.
